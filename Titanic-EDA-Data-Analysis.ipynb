```python
# TITANIC DATASET - EXPLORATORY DATA ANALYSIS
# Tools: Pandas, Matplotlib, Seaborn
# Dataset Link: https://www.kaggle.com/c/titanic/data?select=train.csv

# STEP 1: IMPORT LIBRARIES
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Show plots inline in notebook
%matplotlib inline
```

```python
# STEP 2: LOAD THE DATASET
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ETL AND PYTHON/train .csv")  # Make sure train.csv is in the same folder or give correct path
```

```python
# STEP 3: BASIC INFO ABOUT DATA
print("Data Shape:", data.shape)  # rows and columns
print("\nData Types and Null Values:")
print(data.info())  # datatypes and missing values
print("\nFirst 5 rows:")
print(data.head())
```

```python
# STEP 4: STATISTICAL SUMMARY
print("\nDescriptive Stats:")
print(data.describe())  # mean, std, min, max for numerical columns
```

```python

# STEP 5: MISSING VALUE ANALYSIS
print("\nMissing Values Count:")
print(data.isnull().sum())  # Check how many missing values in each column
```

```python
# STEP 6: UNIVARIATE ANALYSIS (SINGLE VARIABLE)
# Barplot for Survived counts
sns.countplot(x='Survived', data=data)
plt.title('Survival Count')
plt.show()
```

```python

# Histogram of Age
sns.histplot(data['Age'].dropna(), kde=True, bins=30)
plt.title('Age Distribution')
plt.show()
```

```python
# STEP 7: BIVARIATE ANALYSIS (TWO VARIABLES)
# Survival vs Sex
sns.countplot(x='Sex', hue='Survived', data=data)
plt.title('Survival by Gender')
plt.show()
```

```python

# Boxplot - Age vs Pclass
sns.boxplot(x='Pclass', y='Age', data=data)
plt.title('Age Distribution across Classes')
plt.show()
```

```python
# STEP 8: CORRELATION ANALYSIS
# Correlation matrix
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python

# STEP 9: PAIRPLOT (Only few columns to avoid overload)
sns.pairplot(data[['Survived', 'Age', 'Fare', 'Pclass']])
plt.suptitle('Pairplot - Selected Features', y=1.02)
plt.show()
```

```python
# STEP 10: FINDINGS SUMMARY
print("""
Key Insights:
1. Females had higher survival rate than males.
2. Younger passengers and passengers in 1st class had better chances of survival.
3. Missing Age values can be imputed using median per class.
4. Fare and Pclass are negatively correlated (rich people in 1st class paid more).
5. Heatmap shows correlation between variables - Pclass, Fare, Survived etc.
""")
```
